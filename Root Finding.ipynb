```python
def bisection_method(f, a, b, iterations):
    steps = []
    for i in range(iterations):
        midpoint = (a + b) / 2
        f_a = f(a)
        f_mid = f(midpoint)
        steps.append((a, b, midpoint, f_a, f_mid))
        
        if f_a * f_mid < 0:
            b = midpoint
        else:
            a = midpoint
            
    return steps
```

```python
def estimated_error(a, b, n):
    return (b - a) / (2**n)
```

```python
iterations = 3

interval_1 = [-1.1, -0.9]
interval_2 = [2.8, 3.0]

steps_interval_1 = bisection_method(f, interval_1[0], interval_1[1], iterations)
steps_interval_2 = bisection_method(f, interval_2[0], interval_2[1], iterations)

results_interval_1 = [(a, b, midpoint, f_mid, estimated_error(a, b, i+1)) for i, (a, b, midpoint, _, f_mid) in enumerate(steps_interval_1)]
results_interval_2 = [(a, b, midpoint, f_mid, estimated_error(a, b, i+1)) for i, (a, b, midpoint, _, f_mid) in enumerate(steps_interval_2)]
```

```python
def f1(x):
    return -0.9*x**2 + 1.7*x + 2.5

def df1(x):
    return -1.8*x + 1.7
```

```python
def newton_raphson(f, df, x0):
    tolerance=1e-4
    max_iterations=100
    x_n = x0
    for i in range(100):
        f_x_n = f(x_n)
        df_x_n = df(x_n)
        x_n1 = x_n - f_x_n / df_x_n
        if abs(x_n1 - x_n) < tolerance:
            return round(x_n1, 4), i+1  
        x_n = x_n1
    return None

x0_1 = -1
x0_2 = 2.8

root_1, iterations_1 = newton_raphson(f1, df1, x0_1)
root_2, iterations_2 = newton_raphson(f1, df1, x0_2)

print(f"Root for f(x) = -0.9x^2 + 1.7x + 2.5, starting at x0 = -1: {root_1}, Iterations: {iterations_1}")
print(f"Root for f(x) = -0.9x^2 + 1.7x + 2.5, starting at x0 = 2.8: {root_2}, Iterations: {iterations_2}")
```

```python
def f2(x):
    return np.tan(x) - 1.2*x

def df2(x):
    return (1 / np.cos(x))**2 - 1.2

root_3, iterations_3 = newton_raphson(f2, df2, 1)
print(f"Root for f(x) = tan(x) - 1.2x, starting at x0 = 1: {root_3}, Iterations: {iterations_3}")
```

```python
root_4, iterations_4 = newton_raphson(f2, df2, -1)
print(f"Root for f(x) = tan(x) - 1.2x, starting at x0 = 1: {root_4}, Iterations: {iterations_3}")
```

```python
def my_fixed_point(f, g, tol, max_iter, a, b):
    def F(x):
        return f(x) - g(x)
    for i in range(max_iter):
        mid = (a + b) / 2
        F_mid = F(mid)
        
        if abs(F_mid) < tol:
            return round(mid, 4)
        
        if F(a) * F_mid < 0:
            b = mid  # The root is in the left half
        else:
            a = mid  # The root is in the right half
    
```

```python
def f(x):
    return -0.9*x**2 + 1.7*x + 2.5

def g(x):
    return x + 2

X = my_fixed_point(f, g, 1e-4, 100, -2, 2)
print('Fixed point:',X)
Y = my_fixed_point(f, g, 1e-4, 100, 1, 3)
print('Fixed point:',Y)
```

```python
x_values = np.linspace(-2,2,1000)
plt.figure(figsize=(12,8))
plt.plot(x_values,f(x_values),label='-0.9x^2 + 1.7x + 2.5')
plt.plot(x_values,g(x_values),label='x+2')
plt.plot([X,Y],[f(X),f(Y)],'x',color='red',label = 'Points of Intersection')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True,linestyle='--',alpha=0.5)
plt.legend()
```
