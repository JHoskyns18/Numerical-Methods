```python
x = np.linspace(-4,4,100)
y = np.linspace(-3,3,100)

X, Y = np.meshgrid(x, y)
Z = (X*Y*(X**2 - Y**2)) / (X**2 + Y**2)

fig = plt.figure(figsize=(14, 6))

ax1 = fig.add_subplot(121, projection='3d')
surf = ax1.plot_surface(X, Y, Z, cmap='viridis')
ax1.set_title('3D Surface Plot of Z(x, y)')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')
ax1.set_zlabel('Z-axis')
fig.colorbar(surf, ax=ax1, shrink=0.5)

ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_wireframe(X, Y, Z, color='purple')
ax2.set_title('3D Wireframe Plot of Z(x, y)')
ax2.set_xlabel('X-axis')
ax2.set_ylabel('Y-axis')
ax2.set_zlabel('Z-axis')

plt.tight_layout()
plt.show()
```

```python
def my_polygon(n):
    angles = np.linspace(0, 2 * np.pi, n + 1) 
    
    x = np.cos(angles)
    y = np.sin(angles)

    plt.figure(figsize=(5, 5))
    plt.plot(x, y, 'b-', marker='o')
    
    plt.title(f'Regular Polygon with {n} Sides')
    plt.xlabel('X')
    plt.ylabel('Y')
    
    plt.axis('equal')
    
    plt.grid()
    plt.show()
```

```python
my_polygon(28)
```

```python
def my_fun_plotter(f, x):
    y = f(x)
    plt.figure(figsize=(6, 4))
    plt.plot(x, y, 'r-', label='f(x)') 
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Plot of the Function f(x)')
    plt.grid()
```

```python
def my_poly_plotter(n, x):
    plt.figure(figsize=(8, 6))
    
    for k in range(1, n + 1):
        y = x ** k
        plt.plot(x, y, label=f'$x^{k}$')
    
    plt.xlabel('x')
    plt.ylabel('p_k(x)')
    plt.title(f'Polynomials p_k(x) = $x^k$ for k = 1 to {n}')
    plt.legend(loc='best')
    plt.grid(True)
```

```python
my_poly_plotter(5,np.linspace(-1, 1, 200))
```

```python
def my_sierpinski(n):
    P1 = np.array([0, 0])
    P2 = np.array([0.5, np.sqrt(3) / 2])
    P3 = np.array([1, 0])
    points = [P1]
    
    for i in range(1, n):
        current_point = points[-1]
        chosen_vertex = random.choice([P1, P2, P3])
        new_point = (current_point + chosen_vertex) / 2
        points.append(new_point)
        
    points = np.array(points)
    plt.plot(points[:, 0], points[:, 1], 'b.', markersize=1)
    plt.axis('equal')
```

```python
my_sierpinski(100)
```

```python
def my_surface_plotter(f, x, y, option):
    X, Y = np.meshgrid(x, y)
    Z = f(X, Y)
    fig = plt.figure()
    if option == 'surface':
        ax = fig.add_subplot(111, projection='3d')
        ax.plot_surface(X, Y, Z, cmap='viridis')
        ax.set_title('3D Surface Plot')
        ax.set_xlabel('X')
        ax.set_ylabel('Y')
        ax.set_zlabel('f(X, Y)')
    
    elif option == 'contour':
        ax = fig.add_subplot(111)
        contour = ax.contourf(X, Y, Z, cmap='magma')
        fig.colorbar(contour)
        ax.set_title('Contour Plot')
        ax.set_xlabel('X')
        ax.set_ylabel('Y')
    
    else:
        raise ValueError("Invalid option. Choose 'surface' or 'contour'.")

    plt.show()
```

```python
f = lambda x, y: np.cos(y)*np.sin(np.exp(x))
my_surface_plotter(f, np.linspace(-1, 1, 20), np.linspace(-2, 2, 40), 'surface')
my_surface_plotter(f, np.linspace(-1, 1, 20), np.linspace(-2, 2, 40), 'contour')
```
