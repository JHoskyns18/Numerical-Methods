## Vectorized drawing of random numbers using numpys' random module is much more efficient because all the numbers are drawn at once.

Remember the definition of mean:
$$
    x_m = \frac{1}{N} \sum_{j=0}^{N-1} x_j   
$$

and of the standard deviation:
$$
    x_s = \sqrt(x_v) \rightarrow x_v=\frac{1}{N}\sum_{j=0}^{N-1} (x_j-x_m)^2
$$

```python
import random
import numpy as np

N = 1000
values = []
means = []
stds = []

for i in range(N):
    x = random.uniform(-1, 1)  # Draw a random number
    values.append(x)  # Store it
    mean = np.mean(values)  # Calculate the mean of all drawn numbers
    std = np.std(values)    # Calculate the std deviation of all drawn numbers
    means.append(mean)  # Store the mean
    stds.append(std)    # Store the std deviation
    print(f"Draw {i + 1}: Mean = {mean:.5f}, Std Dev = {std:.5f}")

# Optionally, at the end you can print the final results:
print("\nFinal Mean:", means[-1])
print("Final Std Dev:", stds[-1])
```

the fraction of the throws the die shows six eyes

```python
def six_eye(N):
    M = 0
    for i in range(N):
        out = random.randint(1,6)
        if out == 6:
            M += 1
    return M/N    
```

## Monte Carlo Integration
One of the earliest applications of random numbers was numerical computation of integrals, a non-random (deterministic) problem. Computing integrals with random numbers is known as Monte Carlo integration and is one of the most powerful techniques throughout science.

The $\textbf {mean-value theorem}$ from calculus states that:
$$
\int_a^b f(x) dx = <f> (b-a)
$$
where $<f>$ is the mean value of $f$, defined as:
$$
<f> = \frac{1}{b-a} \int_a^b f(x) dx
$$

Using numerical methods, we can approximate $<f>$ by taking the average of $f$ at $n$ points $x_0, x_1, x_2, ...x_n$.

$$
<f> \approx \frac{1}{N} \sum_{j=0}^{n-1} f(x_j)
$$

There is now freedom to choose the points $x_0, x_1, ...,x_{n-1}$
We throw random points, uniformly distributed in $[a,b]$

```python
import random
import numpy as np
```

```python
def MonteCarlo_int(f,a,b,n):
    s=0
    for i in range(n):
        x=np.random.uniform(a,b)
        s=s+f(x)
    i_mc=(b-a)/n*s
    return i_mc
```

```python
f = lambda x: np.sin(x)
MonteCarlo_int(f,0,np.pi,100000)
```

```python
import random
import numpy as np

def MC_int_vec(f,a,b,n):
    s=0
    for i in range(n):
        x=np.random.uniform(a,b)
        s=np.sum(f(x))
        I=(b-a)/n*s
    return I    
```

Let's try a Monte Carlo Integration for $f(x) = 2 + 3x$ between [1,2]. Most other numerical integration methods will integrate such a lienar function exactly, regardless of the number of function evaluations. This is not the case with Monte Carlo Integration.

It would be interesting to see how the quality of Monte Carlo approximation increases with $n$. Plot the evolution of the integral (store the intermediate I values).

```python
import numpy as np
import matplotlib.pyplot as plt

def MC_int_test(f, a, b, n):
    np.random.seed(121)
    I = np.zeros(n)  # Store intermediate I values
    s = 0
    for i in range(1,n+1):
        x = np.random.uniform(a, b)
        s += f(x)  # Accumulate the function evaluations
        I[i-1] = (b - a) * s / (i)  # Compute the current estimate of the integral
    
    # Plot the evolution of the integral approximation
    plt.plot(range(n), I, label="Numerical approx")
    plt.xlabel('Number of Samples')
    plt.ylabel('Estimated Integral')
    plt.title('Monte Carlo Integration of f(x) = 2 + 3x')
    plt.grid()
    exact_integral = 6.5
    error=6.5-I
    plt.axhline(y=exact_integral, color='r', linestyle='--', label='Exact Integral = 6.5')
    
    plt.legend()
    plt.show()
    
# Function to integrate
f = lambda x: 2 + 3 * x

# Call the function with the specified parameters
MC_int_test(f, 1, 2, 10000)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def MC_int_test(f, a, b, n):
    np.random.seed(121)  # Set the seed for reproducibility
    estimates = []  # Store all the integral estimates
    for _ in range(10000):  # Repeat 10,000 times
        I = np.zeros(n)  # Store intermediate values for each execution
        s = 0
        for i in range(n):
            x = np.random.uniform(a, b)
            s += f(x)  # Accumulate function evaluations
            I[i] = (b - a) * s / (i + 1)  # Calculate the current estimate of the integral
        estimates.append(I[-1])  # Store the last estimate after n samples
    
    # Display the histogram of estimates
    plt.hist(estimates, bins=30, density=True, alpha=0.6, color='b')
    
    # Calculate the mean and standard deviation of the estimates
    mu = np.mean(estimates)
    sigma = np.std(estimates)
    
    # Create a normal distribution to overlay on the histogram
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-(x - mu)**2 / (2 * sigma**2))
    plt.plot(x, p, 'k', linewidth=2, label='Normal Distribution')
    
    plt.title('Distribution of Monte Carlo Integral Estimates')
    plt.xlabel('Integral Estimate')
    plt.ylabel('Density')
    plt.legend()
    plt.grid()
    plt.show()

# Function to integrate
f = lambda x: 2 + 3 * x

# Call the function with the specified parameters
MC_int_test(f, 1, 2, 1000)
```

## Area Computing by Throwing Random Numbers

We think about some geometric region G in the plane and a surrounding box B with geometry $[x_L,x_H]\times[y_L,y_H]$.

One way to compute the area of G is to draw N random points inside B and count how many of them, M, lie inside of G. The area is then M/N (G's fraction of B's area) times the area of B ($[x_L,x_H]\times[y_L,y_H]$). 

In other words, this method is a kind of dart game where you record how many hits there are inside G if every throw hits uniformly within B.

#### Now we formulate this method for computing the integral $\int_a^b f(x) dx$. We know that the integral is the area under the curve $y = f(x)$  and above the $x$ axis between $[a,b]$.  

We introduce a rectangle $B$:
$$
B = \{(x,y)\} |   a \leq x \leq b, 0 \leq y \leq m
$$

where $m \leq max_{x\epsilon[a,b]} f(x)$

The area is $ \frac{M}{N} \times m \times (b-a)$

```python
def MCInt_dart(f,a,b,n,m):
    below = 0
    x=np.random.uniform(a,b)
    y=np.random.uniform(0,m)
    if y <= f(x):
        below = below + 1
    area = (below/n)*m*(b-a)
    return area
```

```python
def MCInt_dart_vec(f, a, b, n, m):
    x = np.random.uniform(a, b, n)  # Generate n random x values in [a, b]
    y = np.random.uniform(0, m, n)  # Generate n random y values in [0, m]
    below = np.sum(y < f(x))         # Count how many points are below the curve
    area = (below / n) * m * (b - a)   # Compute the estimated area
    return area
```

## Random walk in One space dimension

Now, we want to simulate a collection of particles that move around in a random fashion. This type of simulation is fundamental in physics as well as in many other sciences to describe many phenomena.

Imagine we have some particles that perform random moves, either to the left or to the right. We may flip a coin to decide the movements of each particle, say head = right and tail = left. Each move is one unit length. Physicists use the term $\bf {random \, walk}$ for this type of motion.

Curiosity: the movement is also known as $\it \bf {drunkard's \, walk}$ (well, for obvious reason)

How can we implement $n_s$ random steps of $n_p$ particles? First, we introduce a coordinate system where all the movements are along the $x$ axis. An array of $x$ values then holds the positions of all particles. We draw random numbers to simulate flipping a coin, say we draw integers 1 and 2, where 1 means head (right) and 2 means tail (left).

```python
import numpy as np
import matplotlib.pyplot as plt

HEAD = 1
TAIL = 2  # We simulate flipping a coin where 1 means right and 2 means left
ns = 1000  # Number of steps for each simulation
n_p = 100   # Number of particles
positions = np.zeros(n_p)

# Simulate coin flips
for p in range(n_p):
    for step in range(ns):
        coin = np.random.randint(1, 3)  # Include both HEAD(1) and TAIL(2)
        if coin == HEAD:
            positions[p] += 1
        else:  # Since we only have HEAD or TAIL, we can use else for TAIL
            positions[p] -= 1
# Generate the histogram
plt.hist(positions, bins=10, alpha=0.5, color='magenta')
plt.xlabel('Position')
plt.ylabel('Frequency')
plt.title('Distribution of Final Positions After Coin Flips')
plt.axis([-ns, ns, 0, plt.ylim()[1]])
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

HEAD = 1
TAIL = 2  # We simulate flipping a coin where 1 means right and 2 means left
ns = 100  # Number of steps for each simulation
np_particles = 10   # Number of particles
positions = np.zeros(np_particles)

# Open a file to save the mean and standard deviation
with open('statistics.txt', 'w') as stats_file:
    # Create a PDF file to save all histograms
    with PdfPages('histograms.pdf') as pdf:
        # Simulate coin flips
        for step in range(ns):
            for p in range(np_particles):
                coin = np.random.randint(1, 3)  # Include both HEAD(1) and TAIL(2)
                if coin == HEAD:
                    positions[p] += 1
                else:  # TAIL
                    positions[p] -= 1

            # Calculate the mean position and standard deviation
            mean_pos = np.mean(positions)
            stdev_pos = np.std(positions)

            # Write the mean and standard deviation to the file
            stats_file.write(f"Step {step + 1}: Mean = {mean_pos:.2f}, Standard Deviation = {stdev_pos:.2f}\n")

            # Draw the histogram of positions
            plt.clf()  # Clear the current figure
            plt.hist(positions, bins=20, alpha=0.7, color='blue', edgecolor='black')
            plt.axvline(mean_pos, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean_pos:.2f}')
            plt.axvline(mean_pos + stdev_pos, color='green', linestyle='dashed', linewidth=1, label=f'SD = {stdev_pos:.2f}')
            plt.axvline(mean_pos - stdev_pos, color='green', linestyle='dashed', linewidth=1)
            plt.title(f'Histogram of Positions at Step {step + 1}')
            plt.xlabel('Position')
            plt.ylabel('Number of Particles')
            plt.legend()
            
            # Save the current figure to the PDF
            pdf.savefig()  # Save the current figure in the PDF
            plt.close()  # Close the figure to free up memory
```

# The Salpeter Mass function

The Salpeter mass function [Salpeter, 1955] is one of the most popular initial mass functions for stars. It is defined as $p(m)dm = const * m^{−\alpha} dm$, where α = 2.3. It means that, given a population of stars in the zero-age main sequence, the probability of having a star of mass m in this population is $p(m) = const * m^{-\alpha}$. 
### Massive stars are significantly less common than light stars.
Assuming that the minimum stellar mass is $M_{min} = 0.1 M_\odot$ and the maximum stellar mass is $M_{max} = 150 M_\odot$, randomly calculate the mass of $10^6$ stars distributed according to the Salpeter initial mass function by using the inverse random sampling technique. Plot the resulting population of stellar masses with an histogram

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
M_min = 0.1   # Minimum mass in solar masses
M_max = 150.0 # Maximum mass in solar masses
alpha = 2.3
num_stars = 10**6

# Normalization constant calculation
def normalization_constant(M_min, M_max, alpha):
    # Calculate the normalization constant
    integral_value = (M_max**(-alpha + 1) - M_min**(-alpha + 1)) / (-alpha + 1)
    return 1 / integral_value

C = normalization_constant(M_min, M_max, alpha)

# CDF Inverse Transformation Sampling
def salpeter_sampling(num_stars, M_min, M_max, alpha):
    u = np.random.uniform(0, 1, num_stars)
    
    # Calculate the mass based on the inverse CDF properly
    m = (M_min**(1 - alpha) + u * (M_max**(1 - alpha) - M_min**(1 - alpha)))**(1 / (1 - alpha))
    return m

# Generate star masses
masses = salpeter_sampling(num_stars, M_min, M_max, alpha)

# Define a function for the theoretical Salpeter IMF
def theoretical_salpeter(m):
    return C * m**(-alpha)

# Generate mass values for the theoretical curve
mass_values = np.logspace(np.log10(M_min), np.log10(M_max), 1000)
theoretical_values = theoretical_salpeter(mass_values)

# Plotting the histogram and the theoretical curve

plt.hist(masses, bins=1000, histtype='step', density=True, edgecolor='black', label='Monte Carlo sampling')
plt.plot(mass_values, theoretical_values, color='red', linestyle='--', label='Theoretical Salpeter IMF $y \\propto x^{-2.3}$')

# Labels and title
plt.title('Comparison of Monte Carlo Sampling and Theoretical Salpeter IMF')
plt.xlabel('Mass (M_sun)')
plt.ylabel('Probability Density')
plt.xscale('log')
plt.yscale('log')
plt.grid()
plt.legend()
plt.show()
```

## Exercise
Generate $10^4$ random numbers distributed according to a Gaussian (sigma=2, centre at 0) with the rejection method, by using the $f(x)$=1,
uniform between $min=-50$ and $max=+50$.

Note that $f(x)$ is not a well-defined distribution function and it cannot be since $f(x)$ > $p(x)$ everywhere, where $p(x)$ is a well-defined probability function (in this case the Gaussian PDF). Thus, $\int_{x_{min}}^x f(x) \,dx$ is a uniform random number but it does not necessarily lie between 0 and 1 

```python
import numpy

def gaussian_pdf(x, mu=0, sigma=2):
    """Probability density function of a Gaussian distribution."""
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)
    
# Parameters for rejection sampling
xmin = -50
xmax = 50
n_samples = 10000
sigma = 2
mu = 0

# Maximum value of p(x) over the acceptance region
max_p = gaussian_pdf(mu, mu, sigma)

# Storage for accepted samples
samples = []

while len(samples) < n_samples:
    # Generate uniform samples from f(x)
    x = np.random.uniform(xmin, xmax)
    # Generate a uniform random number for acceptance criterion
    y = np.random.uniform(0, max_p)
    
    # Acceptance criterion
    if y <= gaussian_pdf(x, mu, sigma):
        samples.append(x)

samples = np.array(samples)

# Plotting results to visualize the distribution
plt.hist(samples, bins=20, density=True, alpha=0.6, color='g', label='Generated Samples')
x = np.linspace(-50, 50, 1000)
plt.plot(x, gaussian_pdf(x, mu, sigma), 'r', label='Gaussian PDF (mu=0, sigma=2)')
plt.xlim(-10, 10)
plt.ylim(0, max_p+.04)
plt.title('Rejection Sampling of Gaussian Distribution')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend(loc="upper right")
plt.show()
```
